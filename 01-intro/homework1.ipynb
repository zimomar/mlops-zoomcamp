```python
import pandas as pd
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
def read_dataframe(url):
    df = pd.read_parquet(url)
    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
    df = df[((df.duration >=1) & (df.duration <=60))]

    categorical = ['PULocationID', 'DOLocationID']
    df[categorical] = df[categorical].astype(str)
    
    return df
```

```python
df_jan = read_dataframe('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')
df_feb = read_dataframe('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
categorical = ['PULocationID', 'DOLocationID']

dv = DictVectorizer()

train_dicts = df_jan[categorical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)
```

```python
X_train
```

Q4. 515.

```python
target = 'duration'
y_train = df_jan[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)
mean_squared_error(y_train, y_pred, squared=False)
```

Q5. 7.64

```python
# Testing model on February 2023 dataset
val_dicts = df_feb[categorical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
y_val = df_feb[target].values
```

```python
y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```
